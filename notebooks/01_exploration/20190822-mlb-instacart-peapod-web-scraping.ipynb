# BEAUTIFUL SOUP

## Instacart 

```python
import pandas as pd
import requests 
import scrapy

from bs4 import BeautifulSoup
from urllib.request import urlopen
```

Change the user agent from beautiful soup to Mozilla so that we can bypass the first level of blocking

```python
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'
headers = {'User-Agent': user_agent}
```

```python
url = "https://www.instacart.com/store/marianos/storefront"
response = requests.get(url,headers=headers)
html = response.content
bs = BeautifulSoup(html, 'html.parser')
```

```python
print(bs.h1)
```
