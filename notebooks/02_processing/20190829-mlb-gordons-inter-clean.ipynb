There seems to be some prices that just say reg. Let's take a look at these a little deeper

```python
gordon_prices.loc[(gordon_prices['main_price'] == 'Reg: ')].head()
```

```python
full_item_desc = gordon_prices.full_item_desc
main_price = gordon_prices.main_price
main_price_new = []
for price, fd in zip(main_price, full_item_desc):
    if price == 'Reg: ':
        try:
            main_price_new.append(re.search("<span>\$(.*)</span></span></div></div></div><span class=\"full-item-name\" | </span><span aria-label=\"each\" style=\"font-size: 13px;\">\\xa0each\',", fd).group(1))
        except:
            main_price_new.append(np.nan)
    else: 
        main_price_new.append(price)  
```

Let's make sure we didn't lose any prices

```python
print(len(gordon_prices))
print(len(main_price_new))
```

```python
gordon_prices['main_price'] = main_price_new
```

```python
gordon_prices
```

## Split up the item column 

### Price Per LB

```python
per_lb = gordon_prices['full_item_desc']
```

```python
per_lb_final = []
per_lb_pattern = r'\d*\.[0-9]{2}\/lb'
for item in per_lb:
    try:
        per_lb_final.append(re.search(per_lb_pattern, item).group())
    except:
        per_lb_final.append(np.nan)
```

```python
per_lb_final
```

```python
gordon_prices['price_per_lb'] = per_lb_final
```

### Measure Words

```python
measure_word_pattern = 'each'
measure_word_list = []
for item in full_item_desc:
    try:
        measure_word_list.append(re.search(measure_word_pattern, item).group())
    except:
        measure_word_list.append(np.nan)
```

```python
gordon_prices['measure_words_main_price'] = measure_word_list
```

### Item Weight/ Count/ Vol

```python
item_weight_count_vol = []
for item in full_item_desc:
    try:
        item_weight_count_vol.append(re.search('aria-label=\"\.(.*)\">\d+', item).group(1))
    except:
        item_weight_count_vol.append(np.nan)
```

```python
gordon_prices['item_weight_count_vol'] = item_weight_count_vol
```
