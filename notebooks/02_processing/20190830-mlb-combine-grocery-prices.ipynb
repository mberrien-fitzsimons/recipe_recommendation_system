# One Large Unique Grocery Price List

```python
import pandas as pd
```

```python
hmart = pd.read_csv('../../data/02_intermediate/grocery_prices_hmart.csv')
marianos = pd.read_csv('../../data/02_intermediate/grocery_prices_marianos.csv')
gordons = pd.read_csv('../../data/02_intermediate/grocery_prices_gordons.csv')
```

```python
print('Number of Groceries in Marianos Dataset: ', len(marianos))
print('Number of Groceries in Gordons Dataset: ', len(gordons))
print('Number of Groceries in HMART Dataset: ', len(hmart))
```

```python
marianos.columns
```

```python
gordons.rename(columns={'prod_aile_count':'prod_aile'}, inplace=True)
```

```python
gordons.columns
```

```python
hmart.rename(columns={'prod_aile_count':'prod_aile'}, inplace=True)
```

```python
hmart.columns
```

```python
grocery_prices_full = pd.concat([gordons, hmart, marianos])
```

```python
len(grocery_prices_full)
```

```python
mask = (grocery_prices_full.duplicated())
```

```python
duplicates = grocery_prices_full.loc[mask]
```

```python
duplicates.sort_values(['product'])
```

```python
grocery_prices_full.to_csv('../../data/02_intermediate/grocery_prices_full.csv')
```
